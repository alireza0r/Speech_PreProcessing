**FFT**</br>
Fast Fourier transform

```python
DatasetLoadDir = '/content/MyDrive/MyDrive/TA_amirkabir/Spring_2022/Final_Prj/Datasets/ShEMO/'
FileDirName = os.path.join(DatasetLoadDir, 'female', 'F07A02.wav')
sr = 44100

fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10,8))

sig, sr = librosa.load(FileDirName, sr)
librosa.display.waveshow(sig, sr=sr, max_points=len(sig), ax=ax[0], marker='.', x_axis='s')

# RMS config
WinLen = int(0.040 * sr) # 40 milisecond
WinHop = WinLen // 2
print('Windows lenght: {:d}, Windows: {:d}'.format(WinLen, WinHop))

# FFT
sig_fft = scipy.fft.fft(sig, n=len(sig))
print(sig_fft.shape)

ax[1].plot(np.abs(sig_fft[:20000]), '-o')
ax[1].set_xlabel('n')
ax[1].set_ylabel('FFT')
```

**FFT and windowing**</br>
Fast Fourier transform

```python
# windowing
WinLen = int(0.040 * sr) # 40 milisecond
WinHop = WinLen // 2
print('Windows lenght: {:d}, Hop: {:d}'.format(WinLen, WinHop))

sig, sr = librosa.load(FileDirName, sr)
sig_len = len(sig)
print(f'Sighnal length: {sig_len}')

NWindow = (sig_len - WinHop) / (WinLen - WinHop)
NWindow = int(np.ceil(NWindow))
print(f'Number of window: {NWindow}')

SigNewLen = (WinLen - WinHop)*NWindow + WinHop
print(f'Signal new length: {SigNewLen}')

sig_zero_pad = np.zeros(shape=(SigNewLen, ))
sig_zero_pad[:sig_len] = sig

#window = scipy.signal.get_window(('hann'), Nx=WinLen)
window = np.ones(shape=(WinLen, ))

frame_list = []
for nwindow in range(NWindow):
  shift = WinLen - WinHop
  frame = sig_zero_pad[nwindow*shift:WinLen + nwindow*shift]

  frame_list.append(frame * window)

frames_np = np.array(frame_list)  
print(frames_np.shape)
```

# **MFCC**</br>
Mel-Frequency Cepstral Coefficients

```python
DatasetLoadDir = '/content/MyDrive/MyDrive/TA_amirkabir/Spring_2022/Final_Prj/Datasets/ShEMO/'
FileDirName = os.path.join(DatasetLoadDir, 'female', 'F07A02.wav')
sr = 44100

fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10,8))

sig, sr = librosa.load(FileDirName, sr)
librosa.display.waveshow(sig, sr=sr, max_points=len(sig), ax=ax[0], marker='.', x_axis='s')

# MFCC config
WinLen = int(0.040 * sr) # 40 milisecond
WinHop = WinLen // 2
print('Windows lenght: {:d}, Windows: {:d}'.format(WinLen, WinHop))

sig_mfcc = librosa.feature.mfcc(librosa.power_to_db(sig), sr=sr, n_mfcc=12, fmax=sr//2)
img = librosa.display.specshow(sig_mfcc, x_axis='frames', ax=ax[1])
fig.colorbar(img, ax=ax[1])
```

```python
from scipy import signal
from scipy.fft import fft, fftshift
import matplotlib.pyplot as plt

window = signal.windows.barthann(51)
plt.plot(window)
plt.title("Bartlett-Hann window")
plt.ylabel("Amplitude")
plt.xlabel("Sample")

plt.figure()
A = fft(window, 2048) / (len(window)/2.0)
freq = np.linspace(-0.5, 0.5, len(A))
response = 20 * np.log10(np.abs(fftshift(A / abs(A).max())))
plt.plot(freq, response)
plt.axis([-0.5, 0.5, -120, 0])
plt.title("Frequency response of the Bartlett-Hann window")
plt.ylabel("Normalized magnitude [dB]")
plt.xlabel("Normalized frequency [cycles per sample]")
```

```python
from scipy import signal
from scipy.fft import fft, fftshift
import matplotlib.pyplot as plt

window = signal.windows.boxcar(51)
plt.plot(window)
plt.title("Rectangular window")
plt.ylabel("Amplitude")
plt.xlabel("Sample")

plt.figure()
A = fft(window, 2048) / (len(window)/2.0)
freq = np.linspace(-0.5, 0.5, len(A))
response = 20 * np.log10(np.abs(fftshift(A / abs(A).max())))
plt.plot(freq, response)
plt.axis([-0.5, 0.5, -120, 0])
plt.title("Frequency response of the Rectangular window")
plt.ylabel("Normalized magnitude [dB]")
plt.xlabel("Normalized frequency [cycles per sample]")
```
